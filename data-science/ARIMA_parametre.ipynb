```python
data = data.set_index('annee')
series = data[['elecRenew']]
```

Recherche des bons parametre ARIMA(p,d,q)

```python
X = series.values
```

```python
# évaluer un modèle ARIMA pour un ordre donné (p,d,q)
def evaluate_arima_model(X, arima_order):
	# préparer l'ensemble de données d'entraînement et de test
	train_size = int(len(X) * 0.8)
	train, test = X[0:train_size], X[train_size:]
	history = [x for x in train]
	# faire des predictions
	predictions = list()

# Dans la boucle for, pour chaque point dans l'ensemble de test, 
# nous ajustons le modèle ARIMA sur l'historique, faisons une prévision, 
# l'ajoutons à la liste des prédictions, puis ajoutons la vraie valeur observée à l'historique.

	for t in range(len(test)):
		model = ARIMA(history, order=arima_order)
		model_fit = model.fit()
		yhat = model_fit.forecast()[0]
		predictions.append(yhat)
		history.append(test[t])
	# calculer l'erreur quadratique moyenne 
	rmse = sqrt(mean_squared_error(test, predictions))
	return rmse

# évalue les combinaisons de valeurs p, d et q pour un modèle ARIMA
def evaluate_models(dataset, p_values, d_values, q_values):
	dataset = dataset.astype('float32')
	best_score, best_cfg = float("inf"), None
	for p in p_values:
		for d in d_values:
			for q in q_values:
				order = (p,d,q)
				try:
					rmse = evaluate_arima_model(dataset, order)
					if rmse < best_score:
						best_score, best_cfg = rmse, order
					print('ARIMA%s RMSE=%.3f' % (order,rmse))
				except:
					continue
	print('Best ARIMA%s RMSE=%.3f' % (best_cfg, best_score))

#  ces lignes définissent les valeurs de p, d et q à tester.
p_values = [0, 1, 2, 4, 6, 8, 10]
d_values = range(0, 3)
q_values = range(0, 3)
warnings.filterwarnings("ignore")
evaluate_models(series.values, p_values, d_values, q_values)
```
